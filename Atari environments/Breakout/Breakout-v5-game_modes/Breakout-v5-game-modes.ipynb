# Vectorize the environment

```python
env = make_atari_env("ALE/Breakout-v5", n_envs=4, seed= 0, mode=32)
env = VecFrameStack(env, n_stack=4) #Stack multiple enviroments to run multiple sets of computation
```

```python
env.reset().shape ## 4 outputs
```

```python
log_path = os.path.join("training", "logs")
```

```python
model = A2C("CnnPolicy", env, verbose=1, tensorboard_log=log_path)
```

# Model Training

```python
model.learn(total_timesteps=200000)
```

```python
model_path = os.path.join("training", "saved_models", "A2C_Breakout_Model")
```

```python
model.save(model_path)
```

# Evaluate policy method

```python
env = make_atari_env("ALE/Breakout-v5", n_envs=1, seed= 0)
env = VecFrameStack(env, n_stack=4) # Stack multiple enviroments to run multiple sets of computation
evaluate_policy(model, env, n_eval_episodes=10, render=True) 

#Not able to render it for some reason 
```
