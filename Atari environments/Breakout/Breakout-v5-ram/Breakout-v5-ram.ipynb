# Evaluate policy method

```python
env = gym.make("ALE/Breakout-ram-v5", render_mode= "human")
# We get a frame here since we have breakout-v5
episodes = 1
for episode in range(episodes+1):
    obs = env.reset()
    terminated = False
    score = 0
    i=0
    while not terminated:
        env.render()
        if i == 0:
            action = model.predict(obs[0])[0]
        else:
            action = model.predict(obs)[0]
        obs, reward, terminated, truncated, info = env.step(action)
        score += reward
        i+=1
        if terminated:
            observation, info = env.reset()
                    
    print(f"Episode:{episode} Score:{score}")

env.close()
```

# Chat gpt code test

```python
# Step 1: Install the required libraries
# !pip install stable-baselines3 gymnasium torch matplotlib

# Step 2: Import the required libraries
import gymnasium as gym
import numpy as np
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common.callbacks import EvalCallback, StopTrainingOnRewardThreshold
import torch as th
import matplotlib.pyplot as plt

# Step 3: Initialize the environment
env = gym.make('ALE/Breakout-v5', obs_type='ram')
env = Monitor(env)  # Wrap the environment with Monitor for evaluation
env = DummyVecEnv([lambda: env])  # Wrap the environment

# Step 4: Define a custom neural network architecture (if needed)
policy_kwargs = dict(
    net_arch=[512, 512, 256,256,128],
    activation_fn=th.nn.ReLU
)

# Step 5: Define the model with tuned hyperparameters
model = PPO(
    'MlpPolicy', 
    env, 
    policy_kwargs=policy_kwargs,
    tensorboard_log="./ppo_breakout_tensorboard/"
)

# Step 6: Define a callback to evaluate the model and potentially stop training early
eval_callback = EvalCallback(
    env, 
    best_model_save_path='./logs/', 
    log_path='./logs/', 
    eval_freq=10000,
    deterministic=True, 
    render=False,
    callback_on_new_best=StopTrainingOnRewardThreshold(reward_threshold=20, verbose=1)
)

# Step 7: Train the model
model.learn(total_timesteps=2_000_000, callback=eval_callback)

# Step 8: Evaluate the model
mean_reward, std_reward = evaluate_policy(model, env, n_eval_episodes=10)
print(f'Mean reward: {mean_reward} +/- {std_reward}')

# Step 9: Save the model
model.save("ppo_breakout_ram")

# Optional: Load the model
# model = PPO.load("ppo_breakout_ram")

# Step 10: Plot the rewards for debugging
results = np.load('./logs/evaluations.npz')
plt.plot(results['timesteps'], results['results'])
plt.xlabel('Timesteps')
plt.ylabel('Reward')
plt.title('Reward over Time')
plt.show()
```
